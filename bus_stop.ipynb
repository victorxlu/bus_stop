```python
# Query URL
#stopurl = ('https://data.etabus.gov.hk/v1/transport/kmb/stop/FE30EA565CC9ADBE')
#url = ('https://data.etabus.gov.hk/v1/transport/kmb/route-stop/39A/outbound/1')  #outbound
url = ('https://data.etabus.gov.hk/v1/transport/kmb/route-stop/203E/inbound/1')  #inbound
#url = ('https://data.etabus.gov.hk/v1/transport/kmb/stop-eta/33674BF8F361D2C3')
#url = ('https://data.etabus.gov.hk/v1/transport/kmb/route-eta/39A/1)

# Some api's will have nicer syntax like:
# `&time=2010..2012` or `&na_item=B1GQ,D21`

response = requests.get(url)  # Make a GET request to the URL

# Print data returned (parsing as JSON)
payload = response.json()  # Parse `response.text` into JSON

# NOTE: Could use print(response.json()) but this wouldn't be formatted nicely
#import pprint

import pprint
pp = pprint.PrettyPrinter(indent=1)
pp.pprint(payload)

print(response.json())
```

```python
dir(response)
```

```python
#print(payload)

bus_dataset = pd.DataFrame(payload['data'], columns=['bound', 'route', 'seq', 'service_type', 'stop'])

# Display the DataFrame
print(bus_dataset)

#print (busdata)
```

```python
import json
import requests
import pprint
import pandas as pd

df1 = bus_dataset

i = len(df1['stop'])

x = 0

stop_dataset = []

for x in range(i):

  stoplist = df1['stop'][x]

  #structure URL
  stopurl = ('https://data.etabus.gov.hk/v1/transport/kmb/stop/'+ stoplist)
  #print(stopurl)

  stopresponse = requests.get(stopurl)  # Make a GET request to the URL

  # Print data returned (parsing as JSON)
  stop_payload = stopresponse.json()  # Parse `response.text` into JSON

  #print(stop_payload["data"]["stop"], stop_payload["data"]["name_en"], stop_payload["data"]["name_tc"], stop_payload["data"]["lat"], stop_payload["data"]["long"])
  dataset = { 'stop' : stop_payload["data"]["stop"], 'en' : stop_payload["data"]["name_en"], 'tc' : stop_payload["data"]["name_tc"], 'stop_lat': stop_payload["data"]["lat"], 'stop_long' :stop_payload["data"]["long"]}
  stop_dataset.append(dataset)

print(stop_dataset)

```

```python
#dataset
bus_df = pd.DataFrame(data=bus_dataset, columns=['bound', 'route', 'seq', 'service_type', 'stop'])

#dataset1
stop_df = pd.DataFrame(data=stop_dataset, columns=['stop', 'en', 'tc', 'stop_lat', 'stop_long'])
```
